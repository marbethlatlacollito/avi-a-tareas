```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")
labels = ['CDIA','CONT','AUTO', 'INFO']
colors = ['#b0c2f2', '#ffda9e', '#fabfb7', '#87cefa']
slices =[120, 80, 30, 20]

plt.pie(slices, labels = labels, colors=colors,
        wedgeprops={'edgecolor':'#000000'})

plt.title('Mi Gráfico de pastel')
plt.tight_layout()
plt.show()

#colores
#Blue = #008fd5
#Red = #fc4f30
#yellow = #e5ae37
#Green = #69904f
```
